# chargement

```python
import pandas as pd
import re

df = pd.read_csv("../dataset/raw/resume_jd_match_raw.csv")
print(f"Chargé: {len(df)} exemples")
print("\nLabels:")
print(df["label"].value_counts())
```

# Nettoyage des données

```python

def clean_text(text):
    text = str(text)
    text = re.sub(r'\s+', ' ', text)  # Supprimer espaces multiples
    text = text.strip()
    return text

df["text_clean"] = df["text"].apply(clean_text)
```

# Supprimer doublons

```python
initial = len(df)
df = df.drop_duplicates(subset=["text_clean"])
print(f"Doublons supprimés: {initial - len(df)}")
```

# Sauvegarder nettoyé

```python
df.to_csv("../dataset/processed/resume_jd_match_cleaned.csv", index=False)
print(f"✅ Nettoyé: {len(df)} exemples")
```

```python
import pandas as pd
df = pd.read_csv("../dataset/processed/resume_jd_match_cleaned.csv")

# Calcule la longueur du texte nettoyé
df["text_length"] = df["text_clean"].str.len()

# Affiche les statistiques (min, max, moyenne)
print(df["text_length"].describe())

# Affiche les 5 lignes avec le texte le plus court pour les inspecter
print(df.nsmallest(5, 'text_length'))
```
